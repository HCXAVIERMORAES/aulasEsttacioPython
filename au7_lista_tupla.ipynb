```python
#tuplas são listas porem imutaveis(não podemos alterar seus elementos) e devemos respeitar os indices dela.
#Enquanto a lista se usa colchetes "[]" em quanto a tupla se usa parenteses "()"
minha_tupla = ('t','u','p','l','a')        #criação de tupla
minha_tupla                                #acessando a tupla
minha_tupla [1]                            # acessando a posição 1 da tupla

#Existe uma forma de acessar os elementos de uma lista que também serve para as tuplas, chamada de slice, ou seja o 1ª numero (indice)
#até o penultima posição do segundo indice passado, ou seja [2], já que se passou o indice [3]
#O slice nos permite fatiar uma tupla (ou lista) e retornar seus elementos por meio de outra tupla (ou lista), mas sem retirá-los de lá.
#A fatia é representada pelos colchetes, que se referem aos índices da tupla.Índices não existentes geram a exceção “out of range”.
minha_tupla[1:3]      #slice
minha_tupla[2:]       #pega os elementos a partir da posição 2
minha_tupla[:3]       #pega os n primeiros elementos, neste caso os 3 prim.
minha_tupla[:]        #pega a tupla toda

#. VERIFICAR QUANTOS ELEMENTOS EXISTEM NA TUPLA ,metodo len():
len(minha_tupla)
len(("Olá","mundo","do","Python"))

#Concatenar tuplas: usa-se o operador '+'
parte1=("Ouviram","do","Ipiranga")
parte2=("as","margens","plácidas")
verso1 = parte1+parte2
verso1

#Repetir elementos, usa-se o operador '*'
mi=('mi',)         #Note o detalhe da vírgula!
mi*10
#Quando definimos uma tupla com um único elemento, devemos inserir a vírgula, como fizemos aqui. Se apenas criarmos uma tupla com um único elemento 
# a linguagem entenderá que se trata de uma string entre parênteses
mi2 = ('mi')
mi2*10

#verificar se um elemento pertence à tupla, o que vale para a lista tbm vale p tupla
'u' in minha_tupla      #resp true
'f' in minha_tupla      #resp false

#Interar nos elementos, a partir do comando for(repetir trechos de códigos), tbm pode ser usado em listas, arrays e tuplas, tbm chamdo foreach
#for i in minha_tupla:
       # print(i)

#Exercicíos
seila = {}
seila[1] = 1
seila['1'] = 2
seila[1] += 1

soma = 0
for k in seila:
        soma += seila [k]

print(soma)
print(seila])
```

```python
#dicionario é um tipo de estrutura de dados em que há mapeamento entre uma chave (key) e um valor (value).
#criar os dicionarios se usam as chaves'{}'
contatos={}
contatos       #mostra o dicionário

#populando-o inserindo um par de chave-valor
contatos['Felix'] = ["1111-2222"]
contatos
#Também podemos fazer isso usando uma lista como valor
#Aqui, criamos uma lista chamada tel_brito e atribuímos os valores a ela (dois telefones). Inserimos a lista na chave “Brito”
#e mostramos o dicionário em seguida.
tel_brito = ["2221-3321","3322-3333"]
contatos["Brito"] = tel_brito
contatos
#continuando a inserir no dicionario
contatos["Piazza"]=["4444-4441"]
contatos["Carlos Alberto"] = ["4444-5541","7777-4545"]
tel_clodoaldo = ["3333-9999","8888-3333","8888-3434"]
contatos["Clodoaldo"]=tel_clodoaldo
contatos  #cada par chave-valor tem um formato padrão, em que há a chave entre aspas, dois pontos (“:”) e, em seguida, seu valor 

#Acessar dados no dicionario. Sempre usando chave-valor
tel_capitao = contatos["Carlos Alberto"]       #criamos uma variável para receber
tel_capitao
contatos["Felix"]                              # modo mais simples de acesso  

#Aterar o conteúdo de alguma chave no dicionario
contatos["Felix"]
contatos["Felix"] = ["9999-1010"]
contatos["Felix"]

#Métodos para ordenar dicionário Uma alternativa é ordenar uma lista com as chaves e acessar os valores na ordem dessa lista.
# Vejamos alguns métodos para isso.
#Método keys() Permite identificar as chaves de um dicionário
contatos.keys()

#Método values() Permite visualizar os valores armazenados. Ambos os métodos retornam listas
contatos.values()

#Método get() Permite obter o conteúdo associado à chave. Se a chave não existir, o valor associado retornará. Mas, se o valor 
#não for passado, as chamadas do get para chaves inexistentes retornarão um valor denominado None (“nada”).
contatos
contatos.get('Brito')  
contatos.get('Pele')   #não há valor armazenado para a chave ‘Pele’

#Método in Permite verificar se uma chave pertence a um dicionário
'Brito' in contatos          #retorna true
'Pele' in contatos           #false

#Método items() Permite retornar os elementos na forma de tuplas:
contatos.items()

# Método copy() Permite copiar um dicionário:
contatos
jogadores = contatos.copy()
jogadores

#Método clear() Permite apagar todos os elementos do dicionário:
jogadores.clear()
jogadores

#exercicio - Analise o código a seguir:
init_tuple = ()
print (init_tuple.__len__())
#O que será impresso 

#Outras formas de criar dicionários. Também podemos criar um dicionário de outras formas. Uma delas é pelo método dict(), que pode receber dois parâmetros:
#Listas de tuplas: Cada uma contém uma chave e um valor.   ; Sequência de itens: No formato chave-valor.
camisas = dict([(7,"Jairzinho"),(9,"Tostao"),(10,"Pele"),(11,"Rivelino")])
camisas[7]
camisas[10]
#Nesse caso, usamos tuplas, mas poderíamos ter feito o mesmo processo com listas.

#Usando o método fromkeys(lista,valor), é possível criar um dicionário, cujas chaves são os elementos de uma lista, e cujos valores são
#todos iguais a determinado valor. Se o valor não for especificado, o default é None.
{}.fromkeys([16,4,2,3])  #criamos um dicionário com as chaves 16,4,2,3 pertencentes a uma lista. Não definimos nenhum valor a elas.

#melhorando o exemplo anterior:
{}.fromkeys([16,4,2,3],"defesa")  #Aqui, criamos um dicionário com as chaves 1,4,2,3 com o valor “defesa”
```

```python
#Atividades
#1. Nosso DNA é composto de nucleotídeos que contêm as bases nitrogenadas, representados por quatro letras:
#A = adenina;
#C = citosina;
#T = timina;
#G = guanina.
#Considere a sequência de DNA demonstrado por uma tupla: “AATCGGGTTAATGCAATTCG”
#DNA = (‘a’,’a’,’t’, ’c’, ’g’, ’g’, ’g’, ’t’, ’t’, ’a’, ’a’, ’t’, ’g’, ’c’, ’a’, ’a’, ’t’, ’t’, ’c’, ’g’)
#Agora, faça um programa ou script em Python para identificar a quantidade de timinas (T) que existem nessa sequência.
DNA = ('a','a','t','c','g','g','g','t','t','a','a','t','g','c','a','a','t','t','c','g')
t=0
for i in DNA:
  if i=='t':
    t+=1
print(t)

#2. Usando a mesma sequência de DNA, faça um programa ou script para contar cada base.
a=0
c=0
t=0
g=0
for i in DNA:
  if i == 'a':
    a+=1
  elif i == 'c':
    c+=1
  elif i =='t':
    t+=1
  else:
    g+=1

#deve-se transformar a numero inteiro em string "str()" pois so se concatena entre iguais
print("a = "+str(a), " c = "+str(c)," t= "+str(t), " g= "+str(g)) 

```
