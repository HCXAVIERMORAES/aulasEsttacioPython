```python
#Aula 8 Manipulação de dados
#Leitura e escrita de arquivos-texto. Basicamente, ler e escrever arquivos significa trabalhar com strings, armazená-las e recuperá-las. 
#como em outras linguagens de programação, usamos os métodos (ou as funções) open e write para realizar esses processos.
#Abertura de arquivos: usamos o comando open(), com o auxilio de uma variavel,que recebe dois argumentos: caminho(path) e Nome do arquivo e Modo de leitura
#arquivo 'ajax_info.txt” foi aberto para escrita e atribuído à variável “a”. Entre os modos usados para ler e escrever em arquivos, destacamos:
#- r: Abre o arquivo apenas para leitura (read). Você não pode inserir ou modificar o conteúdo ; - w: Abre o arquivo apenas para escrita. O arquivo que 
#existia será sobrescrito pelo atual ; - a–append: Abre o arquivo para inserção de dados que serão escritos no final do arquivo; 
#- r+ : Abre o arquivo para leitura e escrita.  O open(), pode-se omitir o modo. Nesse caso, o padrão será o modo “r”.

#os arquivos são temporarios, para deixa-los permanetes deve-se usar o googleDrive

#a = open('C:\\caminho\\ajax_info.txt', 'r')
#a  #mostra o conteudo da variavel que recebeu o open()
#<_io.TextIOWrapper name='c:\\Users\\Fabiano\\ajax_info.txt' mode='w' encoding='cp1252'> informação que é retornado no IDLE

#Para abrir o arquivo, use a open()função embutida. A open()função retorna um objeto de arquivo, que
#tem um read() método para ler todo o conteúdo do arquivo:
#Abri e leia o arquivo apos anexa-lo. É uma boa prática sempre fechar o arquivo quando terminar de usá-lo.
f = open("demofile2.txt", "r")
print(f.read())
f.close()
#Por padrão, o método read() retorna todo o texto, mas você também pode especificar quantos caracteres deseja retornar:
#Retorne os 5 primeiros caracteres do arquivo.
f = open('demofile2.txt','r')
print(f.read(5))
f.close()
#f = open("demofile2.txt", "a")
#f.write("Agora o arquivo tem mais conteudo!")
#f.close()

#Metodos para manipulação de arquivos
#ITERAR EM UM ARQUIVO: percorrer o arquivo aberto linha a linha. Usamos o comando for.
a = open('ajax.txt','r')  #ou a = open('ajax.txt')
for linha in a:
  print(linha)
a.close()

#Em outras situações, é melhor ler o conteúdo em listas. Nesse caso, o método a ser usado é o readlines().
a = open("ajax.txt")
lista = a.readline() #retorna uma linha 
lista = a.readlines() #retorna uma lista [] com todas as linhas
lista

#Todo o conteúdo do arquivo será inserido em uma lista. Após fazer isso, os métodos de procura podem ser
#usados para buscar um valor específico ou outro tipo de processamento.

#Apesar de ser traduzido como “abrir”, o método open() pode ser usado para abrir ou criar um arquivo vazio. 
#Para isso, passamos como parâmetro o modo “w”.porem, Quando executamos esse comando, se houvesse um arquivo 
#com o mesmo nome que passamos por parâmetro, o arquivo seria apagado e se criaria um novo sobre o antigo
b=  open('novoarquivo.txt','w')

#Escrever em arquivo: A leitura de dados é feita usando o read(). Para escrever em um arquivo-texto, 
#usam o método write().mos
b.write('Ouviram do ipiranga as margens plácidas') #escreve e retorna o nª de  carracter digitado
b.close()  #fecha o arqivo

c = open('novoarquivo.txt','r')
for linha in c:
  print(linha)
c.close()

#ADICIONAR MAIS CONTEÚDO AO ARQUIVO EXISTENTE: Normalmente, abrimos o arquivo na primeira linha.
a = open('/content/ajax.txt')  # não passamos o parâmetro do modo, logo a linguagem assume que é o modo “r”. 
dados = a.readlines()  #Atribuímos os dados do arquivo à lista dados.
dados

dados.append("<p>Nova linha inserida</p>")
a = open('/content/ajax.txt','w')  #inseri-se nova linha
a.writelines(dados)  # Abri-se o arquivo em modo de escrita e usa-se o método writelines() contendo os dados antes
                     #armazenado
#O método writelines() é diferente do write() porque aceita uma lista de strings, 
#enquanto o write aceita uma string apenas.
```

```python
#STRINGS: em Python, tudo é objeto – inclusive a strings, que deriva da classe str.
#Algumas funções relacionadas às strings: Acessar os caracteres da string por seu índice pois elas
#são arrays de caracteres.Mas há um detalhe importante: as strings são imutáveis quanto à alteração por índice.
str1 =  'Nabucodonosor'
str1
str1[1]

#3Iterar em strings: Como um array ou uma lista, podemos iterar na string com o comando for
for i in str1:
  print(i)

#também pode fatiar (slice) a string
str1[0:5]    #'Nabuc'
str1[:5]     #['Nabuc']
str1[5:]     #'odonosor'
str1[-1]     #'r'
str1[:]      #'Nabucodonosor'

#Verificar se um caractere está na string
'f' in str1   #false
'n' in str1   #true
#Identificar o tamanho da string
len(str1)     #13
#Converter a string para maiúsculas ou minúsculas
str1.upper()     #'NABUCODONOSOR'
str1.lower()     #'nabucodonosor'

#Converter um número para uma string
x = 666
type(x)      #<class 'int'>
x1 = str(x)
type(x1)     #<class 'str'

#Verificar se a string contém caracteres não alfabéticos
str1.isalpha()  #True
str1="Nabucodonosor123"
str1.isalpha()   #false

#Retirar espaços em branco no início e no fim da string: Este método não modifica o conteúdo da string.
str2 = " Nabucodonosor "
str2      #' Nabucodonosor '
str2.strip()

#Juntar cada item da string por meio de um delimitador especificado: o método não vai alterar a string:
str2
"-".join(str2)  #'-N-a-b-u-c-o-d-o-n-o-s-o-r-' 
"-".join(str1)  #'N-a-b-u-c-o-d-o-n-o-s-o-r-1-2-3'
```
